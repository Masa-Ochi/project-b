```python
def showData(ticker, startDate, endDate):

    # 株価チャートの描画
#     df_price = df.query("ticker==@ticker")\
#                 [['date', 'open', 'high', 'low', 'close', 'volume']]\
#                 .set_index("date")\
#                 .rename(columns={
#                     'open'  : 'Open',
#                     'high'  : 'High',
#                     'low'   : 'Low',
#                     'close' : 'Close',
#                     'volume': 'Volume',
#                     })
#     mpf.plot(df_price.query("date > @startDate & date < @endDate"), volume=True, type='candle', style='yahoo', figratio=(12,4))

    # Revenue&Profitの表を描画
    Revenue_and_Profit = [d.get('INDICATOR') for d in macrotrends_scrapying_rules[0]["INDICATORS"]]
    display(df_macrotrends.query("SYMBOL==@ticker & INDICATOR in @Revenue_and_Profit")[["DATE", "INDICATOR", "VALUE"]]\
                    .set_index("DATE").query("DATE > @startDate & DATE < @endDate")\
                    .pivot_table(index="DATE", columns="INDICATOR", values="VALUE").T)
    print("--------------------------------------------------------------------------------------------------------------------------------")
    # Assets&Liabilitiesの表を描画
    Assets_and_Liabilities = [d.get('INDICATOR') for d in macrotrends_scrapying_rules[1]["INDICATORS"]]
    display(df_macrotrends.query("SYMBOL in ('BA') & INDICATOR in @Assets_and_Liabilities")[["DATE", "INDICATOR", "VALUE"]]\
                    .set_index("DATE").query("DATE > @startDate & DATE < @endDate")\
                    .pivot_table(index="DATE", columns="INDICATOR", values="VALUE").T)
    print("--------------------------------------------------------------------------------------------------------------------------------")
    # Marginsの表を描画
    Margins = [d.get('INDICATOR') for d in macrotrends_scrapying_rules[2]["INDICATORS"]]
    display(df_macrotrends.query("SYMBOL in ('BA') & INDICATOR in @Margins")[["DATE", "INDICATOR", "VALUE"]]\
                    .set_index("DATE").query("DATE > @startDate & DATE < @endDate")\
                    .pivot_table(index="DATE", columns="INDICATOR", values="VALUE").T)
    print("--------------------------------------------------------------------------------------------------------------------------------")
    # PriceRatiosの表を描画
    Price_Ratios = [d.get('INDICATOR') for d in macrotrends_scrapying_rules[3]["INDICATORS"]]
    display(df_macrotrends.query("SYMBOL in ('BA') & INDICATOR in @Price_Ratios")[["DATE", "INDICATOR", "VALUE"]]\
                    .set_index("DATE").query("DATE > @startDate & DATE < @endDate")\
                    .pivot_table(index="DATE", columns="INDICATOR", values="VALUE").T)
    print("--------------------------------------------------------------------------------------------------------------------------------")
    # OtherRatiosの表を描画
    Other_Ratios = [d.get('INDICATOR') for d in macrotrends_scrapying_rules[4]["INDICATORS"]]
    display(df_macrotrends.query("SYMBOL in ('BA') & INDICATOR in @Other_Ratios")[["DATE", "INDICATOR", "VALUE"]]\
                    .set_index("DATE").query("DATE > @startDate & DATE < @endDate")\
                    .pivot_table(index="DATE", columns="INDICATOR", values="VALUE").T)
    print("--------------------------------------------------------------------------------------------------------------------------------")
```
