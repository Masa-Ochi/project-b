# 前処理

## 欠損値の除去

```python
df = df[~df[["VALUE"]].isnull().any(axis=1)]
```

```python
df.shape
```

```python
df["SYMBOL"].nunique()
```

## VALUEの数値化

```python
df = df.assign(VALUE_NUMERIC = df["VALUE"].apply(lambda x: x.replace(",", "") if type(x) is str else x).astype("float"))
```

```python
df = df[["SYMBOL", "INDICATOR", "DATE", "VALUE_NUMERIC", "UNIT"]]
df = df.rename(columns={'VALUE_NUMERIC': 'VALUE'})
```
